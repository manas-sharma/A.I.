# Linear regression with one variable:

### Load Data 

```python
import numpy as np

data1 = np.loadtxt('./ex1/ex1data1.txt', dtype='float', delimiter=',')
n1 = len(data1)
```

### Plotting Data

```python
import matplotlib.pyplot as plt

X = np.reshape(data1[:,0],(n1,1))
y = np.reshape(data1[:,1],(n1,1))

plt.title('Scatter plot of training data')
plt.scatter(X,y)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

### Computing Initial Cost

```python
def computeCost(X,y,theta,n):
    hx = np.matmul(X,theta)
    J = np.sum(np.power(hx-y, 2))/(2*n)
    return J
```

```python
X = np.append(np.ones((n1,1)),X,axis=1)
theta = np.zeros((2,1))
print('Initial value of Cost Function: ',computeCost(X, y, theta, n1))
```

### Optimization (Gradient Descent) and Final Cost

```python
def gradientDescent(X,y,theta,learningRate,maxIterations, n):
    cost = np.zeros(maxIterations)
    for i in range(maxIterations):
        hx = np.matmul(X,theta)
        temp = hx - y
        theta = theta - (learningRate/n1) * np.dot(X.T, temp)
        cost[i] = computeCost(X, y, theta, n1)
    return cost, theta
```

```python
maxIterations = 2000
learningRate = 0.01
cost, theta = gradientDescent(X,y,theta,learningRate,maxIterations, n1)
print('Final value of Cost Function: ',cost[-1])
```

### Plotting Data and the Model

```python
X1 = np.reshape(data1[:,0],(n1,1))
y1 = np.reshape(data1[:,1],(n1,1))
plt.scatter(X1,y1,label='Traning Data')
plt.plot(X[:,1],np.matmul(X,theta),c='magenta',label='Predicted Model')
plt.title('Training data with linear regression Ô¨Åt')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.legend()
plt.show()
```

### Plotting the Cost function

```python
plt.plot(np.arange(maxIterations),cost)
plt.title('Cost function values over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

# Linear regression with multiple variables

### Load Data

```python
data2 = np.loadtxt('./ex1/ex1data2.txt', dtype='float', delimiter=',')
n2 = len(data2)
```

### Feature Normalization

```python
mu = np.mean(data2, axis=0)
std = np.std(data2, axis=0)
data2 = (data2 - mu)/std
X = np.reshape(data2[:,:-1],(n2,2))
y = np.reshape(data2[:,-1],(n2,1))
```

### Computing Initial Cost

```python
X = np.append(np.ones((n2,1)),X,axis=1)
theta = np.zeros((3,1))
print('Initial value of Cost Function: ',computeCost(X, y, theta, n2))
```

### Optimization (Gradient Descent) and Final Cost

```python
maxIterations = 1500
learningRate = 0.01
cost, theta = gradientDescent(X,y,theta,learningRate,maxIterations,n2)
print('Final value of Cost Function: ',cost[-1])
```

### Plotting the Cost function

```python
plt.plot(np.arange(maxIterations),cost)
plt.title('Cost function values over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```
